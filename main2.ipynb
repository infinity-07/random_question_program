```python
import csv

# 从 CSV 文件中读取数据并重新储存到 questions_list 中
questions_list = []
with open('完整题库.csv', 'r', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # 将字符串转换回列表
        row['options'] = row['options'].split(';')
        # 将chapter字段转换回整数
        row['chapter'] = int(row['chapter'])
        questions_list.append(row)
```

```python
# 输出问题到txt文件中，做联系用

import random

def format_question(question, question_number):
    """
    格式化题目，每个选项换行显示。
    
    参数:
    question (dict): 包含题目内容、选项和答案的字典。
    question_number (int): 题目的编号。
    
    返回:
    str: 格式化后的题目字符串。
    """
    question_text = f"题目 {question_number} ({question['question_type']}): {question['question']}"
    options_text = '\n'.join(question['options'])
    return question_text + '\n' + options_text

def quiz_maker(questions_list, chapter_list=range(18), question_num=1, output_file='quiz_output.txt'):
    """
    创建一个包含指定章节题目的测验，并将结果写入文件。
    
    参数:
    questions_list (list): 包含所有题目的列表。
    chapter_list (list): 指定要包含的章节号列表，默认为 range(18)。
    question_num (int or str): 要输出的题目数量，默认为 1；如果是 'full'，则输出所有题目。
    output_file (str): 输出文件名，默认为 'quiz_output.txt'。
    """
    # 筛选出指定章节的题目
    filtered_questions = [question for question in questions_list if question['chapter'] in chapter_list]
    


    if not filtered_questions:
        print("没有找到题目。")
        return
    
    # 根据 question_num 的值决定输出题目的数量
    if question_num == 'full':
        questions_to_display = filtered_questions
    else:
        questions_to_display = random.sample(filtered_questions, min(question_num, len(filtered_questions)))
    
    # 乱序输出题目
    random.shuffle(questions_to_display)
    
    # 准备文件写入
    with open(output_file, 'w', encoding='utf-8') as file:
        # 存储格式化后的题目
        formatted_questions = [format_question(question, i+1) for i, question in enumerate(questions_to_display)]
        
        # 写入所有题目到文件
        for question in formatted_questions:
            file.write(question + '\n\n')  # 在每道题后添加两个换行符
        
        # 题目之间打印空行，然后打印答案
        file.write("\n答案：\n")
        for i, question in enumerate(questions_to_display):
            file.write(f"题目 {i+1} 的答案是: {question['answer']}\n")

# 输出所有题目的乱序，答案在最后统一输出
quiz_maker(questions_list, chapter_list=[17], question_num='full')
```

```python
import random
import json

def format_question(question, question_number):
    """
    格式化题目，每个选项换行显示。
    
    参数:
    question (dict): 包含题目内容、选项和答案的字典。
    question_number (int): 题目的编号。
    
    返回:
    str: 格式化后的题目字符串。
    """
    question_text = f"题目 {question_number} ({question['question_type']}): {question['question']}"
    options_text = '\n'.join(question['options'])
    return question_text + '\n' + options_text

def quiz_maker(questions_list, chapter_list=range(18), question_num=1):
    """
    创建一个包含指定章节题目的测验，并逐题输出，等待用户输入答案并判断对错。
    
    参数:
    questions_list (list): 包含所有题目的列表。
    chapter_list (list): 指定要包含的章节号列表，默认为 range(18)。
    question_num (int or str): 要输出的题目数量，默认为 1；如果是 'full'，则输出所有题目。
    """
    # 筛选出指定章节的题目
    filtered_questions = [question for question in questions_list if question['chapter'] in chapter_list]
    
    if not filtered_questions:
        print("没有找到题目。")
        return
    
    # 根据 question_num 的值决定输出题目的数量
    if question_num == 'full':
        questions_to_display = filtered_questions
    else:
        questions_to_display = random.sample(filtered_questions, min(question_num, len(filtered_questions)))
    
    # 乱序输出题目
    random.shuffle(questions_to_display)
    
    incorrect_questions = []

    # 逐题输出并判断答案
    for i, question in enumerate(questions_to_display):
        formatted_question = format_question(question, i+1)
        print(formatted_question)
        user_answer = input("请输入您的答案：")
        
        correct_answer = question['answer']
        if user_answer.upper() == correct_answer:
            print("回答正确！\n")
        else:
            print(f"回答错误。正确答案是: {correct_answer}\n")
            incorrect_questions.append(question)
    
    # 保存错误题目到 CSV 文件
    if incorrect_questions:
        with open('错题.csv', 'a', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['chapter', 'question_type', 'question', 'options', 'answer']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            
            for question in incorrect_questions:
                # 将列表转换为字符串存储
                question['options'] = ';'.join(question['options'])
                writer.writerow(question)

quiz_maker(questions_list, chapter_list=[9], question_num='full')
```
